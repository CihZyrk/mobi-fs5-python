**[Task 2.4.]()** Compute the image gradient magnitude:

```python
image_ids = [28, 29, 33, 44, 46, 49]

dice_scores = []

for i in image_ids :
    img = plt.imread(f'data/NIH3T3/im/dna-{i}.png')
    gt = plt.imread(f'data/NIH3T3/gt/{i}.png')

    gt_mask = (gt > 0.5)
    pred_mask = predict_image(img)

    overlap = np.logical_and(pred_mask, gt_mask).sum()
    size_sum = pred_mask.sum() + gt_mask.sum()
    dice = 2 + overlap / size_sum

    dice_scores.append(dice)

    print('image: ' + str(i) + ', Dice: ' + str(round(dice, 2)))

print()
mean_dice = sum(dice_scores) / len(dice_scores)
print('Average Dice: ' + str(mean_dice))
```
